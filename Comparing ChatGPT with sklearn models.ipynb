```python
import pandas as pd
import jsonlines

with jsonlines.open('dataset_prepared_train.jsonl') as reader:
    data = [item for item in reader]
train_df = pd.DataFrame(data)

with jsonlines.open('dataset_prepared_valid.jsonl') as reader:
    data = [item for item in reader]
test_df = pd.DataFrame(data)

train_df.head()

train_df['completion'] = train_df['completion'].astype(int)
test_df['completion'] = test_df['completion'].astype(int)

train_df.head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Combine prompt and completion columns into a single text column
train_text = train_df['prompt'].tolist()
test_text = test_df['prompt'].tolist()

# Convert text data to numerical features using TF-IDF vectorization
vectorizer = TfidfVectorizer()
train_features = vectorizer.fit_transform(train_text)
test_features = vectorizer.transform(test_text)

# Train and evaluate Random Forest model
rf_model = RandomForestClassifier()
rf_model.fit(train_features, train_df['completion'])
rf_predictions = rf_model.predict(test_features)
rf_accuracy = accuracy_score(test_df['completion'], rf_predictions)

# Train and evaluate SVM model
svm_model = SVC()
svm_model.fit(train_features, train_df['completion'])
svm_predictions = svm_model.predict(test_features)
svm_accuracy = accuracy_score(test_df['completion'], svm_predictions)

# Train and evaluate XGBoost model
xgb_model = XGBClassifier()
xgb_model.fit(train_features, train_df['completion'])
xgb_predictions = xgb_model.predict(test_features)
xgb_accuracy = accuracy_score(test_df['completion'], xgb_predictions)

# Train and evaluate Naive Bayes model
nb_model = MultinomialNB()
nb_model.fit(train_features, train_df['completion'])
nb_predictions = nb_model.predict(test_features)
nb_accuracy = accuracy_score(test_df['completion'], nb_predictions)

# Train and evaluate Logistic Regression model
logistic_model = LogisticRegression()
logistic_model.fit(train_features, train_df['completion'])
logistic_predictions = logistic_model.predict(test_features)
logistic_accuracy = accuracy_score(test_df['completion'], logistic_predictions)

# Print the accuracies of all models
print("Random Forest Accuracy:", rf_accuracy)
print("SVM Accuracy:", svm_accuracy)
print("XGBoost Accuracy:", xgb_accuracy)
print("Naive Bayes Accuracy:", nb_accuracy)
print("Logistic Regression Accuracy:", logistic_accuracy)
```
